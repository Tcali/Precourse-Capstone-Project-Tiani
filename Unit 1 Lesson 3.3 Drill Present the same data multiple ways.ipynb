```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Loading an example dataset provided by the seaborn package.
crashes = sns.load_dataset("car_crashes")
crashes.index=crashes['abbrev']
#crashes=crashes.drop(['ins_premium', 'ins_losses'], axis=1)
crashes.head(n=4)
```

```python
sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(15, 15))
ax = sns.boxplot(x='abbrev', y='total',
                 data=crashes, palette='pastel')

plt.title('Boxplot: Total Crashes')

sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='car crashes')

plt.show()
```

```python
sns.set_style("ticks")
g = sns.factorplot(x="abbrev", y="total", data=crashes,
                   size=15, kind="bar", palette="pastel", ci=95)
g.despine(left=False, trim = True)
g.set_ylabels("Crashes")
g.set_xlabels("")
plt.title('Barplot: Total crashes per state')
plt.show()
```

```python
# Setting the overall aesthetic.
sns.set_style("ticks")
g = sns.factorplot(x="abbrev", y="total", data=crashes,
                   size=15, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Total Crashes")
g.set_xlabels("")
plt.title('Pointplot: Total Crashes by State')
plt.show()
```

```python
#Variable
sns.violinplot(data=crashes)
sns.despine(offset=10, trim=True)
```

```python
g = sns.lmplot(y='alcohol', # Variable 1.
               x='ins_premium', # Variable 2.
               data=crashes, # Data
               fit_reg=True, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("Alcohol related crashes")
g.set_xlabels("Insurance premiums")
plt.title('Scatterplot: Alcohol related crashes by insurance premiums')
plt.show()
```

```python
g = sns.lmplot(y='alcohol', # Variable 1.
               x='no_previous', # Variable 2.
               hue='abbrev',
               data=crashes, # Data
               fit_reg=False, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("Alcohol related crashes")
g.set_xlabels("Amount of people who had no previous crash records")
plt.title('Scatterplot: Alcohol related crashes by no previous crash record')
plt.show()
```

```python
sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(15, 15))
ax = sns.boxplot(x='total', y='ins_premium',
                 data=crashes, palette='pastel')

plt.title('Boxplot: Total crashes by insurance premiums')

sns.despine(offset=10, trim=False)
ax.set(xlabel='Amount of car crashes', ylabel='Insurance Premiums')

plt.show()
```

```python
States = ['CA', 'VA']
crashes = crashes[crashes['abbrev'].isin(States)]
sns.set_style("ticks")
g = sns.factorplot(x="alcohol", y="ins_premium", data=crashes,
                   size=10, kind="bar", palette="pastel", ci=95)
g.despine(left=False, trim = False)
g.set_ylabels("Insurance Premiums")
plt.title('Barplot: CA avs VA premium insurance rate by alcohol related crashes')
plt.show()
```
